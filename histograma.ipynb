```python
import matplotlib.pyplot as plt
import numpy as np
# Datos simulados (calificaciones de alumnos)
data = np.random.normal(loc=50, scale=12, size=500) # cambiamos al media a 50, la dispercion a 12 y los alubnos a 500
# Crear histograma
plt.hist(data, bins=20, color='orange', alpha=0.7, edgecolor='green')
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.show()
```
