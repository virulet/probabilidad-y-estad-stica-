```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros de la distribución binomial
n = 10  # número de minutos
p = 0.05  # probabilidad de accidente por minuto
num_simulaciones = 1000  # número de intersecciones simuladas

# Simulación de 1000 intersecciones en periodos de 10 minutos
accidentes_simulados = np.random.binomial(n, p, num_simulaciones)

# Contar frecuencia de accidentes entre 0 y 3
valores_k = np.arange(0, 4)
frecuencias = [np.sum(accidentes_simulados == k) for k in valores_k]
probabilidades_simuladas = np.array(frecuencias) / num_simulaciones

# Calcular la distribución teórica binomial
x = np.arange(0, n+1)
y = stats.binom.pmf(x, n, p)

# Graficar la distribución simulada
plt.figure(figsize=(10, 5))
plt.bar(valores_k, probabilidades_simuladas, alpha=0.6, color='b', label='Simulación')
plt.plot(x, y, 'ro-', label='Distribución Binomial teórica')
plt.xlabel('Número de accidentes en 10 minutos')
plt.ylabel('Probabilidad')
plt.title('Distribución de accidentes en 10 minutos')
plt.xticks(range(0, 11))
plt.legend()
plt.grid()
plt.show()

# Calcular y mostrar el valor esperado y la varianza
media_teorica = n * p
varianza_teorica = n * p * (1 - p)

media_simulada = np.mean(accidentes_simulados)
varianza_simulada = np.var(accidentes_simulados)

print(f"Valor esperado teórico: {media_teorica:.2f}")
print(f"Valor esperado simulado: {media_simulada:.2f}")
print(f"Varianza teórica: {varianza_teorica:.2f}")
print(f"Varianza simulada: {varianza_simulada:.2f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetros
lambda_per_minute = 5  # Tasa de llegada de vehículos por minuto
interval_minutes = 15   # Intervalo total en minutos
lambda_total = lambda_per_minute * interval_minutes  # Tasa total en 15 minutos

# Simulación de la llegada de vehículos en 15 minutos
vehicles = np.random.poisson(lambda_total, size=1000)

# Calcular el valor esperado y la varianza
expected_value = np.mean(vehicles)
variance = np.var(vehicles)

# Graficar la distribución
plt.figure(figsize=(10, 5))
plt.hist(vehicles, bins=range(min(vehicles), max(vehicles) + 1), density=True, alpha=0.6, color='b', edgecolor='black', label='Simulación')

# Graficar la distribución teórica de Poisson
x = np.arange(min(vehicles), max(vehicles) + 1)
plt.plot(x, poisson.pmf(x, lambda_total), 'ro-', label='Distribución Poisson teórica')

# Configuración del gráfico
plt.xlabel('Número de vehículos en 15 minutos')
plt.ylabel('Frecuencia relativa')
plt.title('Distribución de Poisson para la llegada de vehículos')
plt.legend()
plt.grid()
plt.show()

# Interpretación de resultados
print(f"Valor esperado: {expected_value:.2f}")
print(f"Varianza: {variance:.2f}")
```
