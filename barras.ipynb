```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['programacion', 'fisica', 'calculo', 'sistemas']
frequencies = [4, 2, 9, 5]
plt.bar(categories, frequencies, color='blue', alpha=0.3, edgecolor='green')
plt.title('Diagrama de Barras: Alumnos por Materia')
plt.xlabel('Materias')
plt.ylabel('Cantidad de Alumnos')
plt.show()
```
