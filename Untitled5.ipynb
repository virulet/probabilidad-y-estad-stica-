```python
import numpy as np
import matplotlib.pyplot as plt

# Función para aplicar el Teorema de Bayes
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    # Cálculo de la probabilidad total de B
    P_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)

    # Cálculo de la probabilidad posterior de A dado B
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / P_B

    return posterior_A_given_B

# Ejemplo de datos
prior_A = 0.3  # Probabilidad previa de A
likelihood_B_given_A = 0.8  # Probabilidad de B dado A
prior_not_A = 1 - prior_A  # Probabilidad previa de no A
likelihood_B_given_not_A = 0.2  # Probabilidad de B dado no A

# Cálculo de la probabilidad posterior
posterior = bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A)

# Graficar
labels = ['P(A)', 'P(B|A)', 'P(~A)', 'P(B|~A)', 'P(A|B)']
values = [prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A, posterior]

plt.figure(figsize=(10, 6))
plt.bar(labels, values, color=['blue', 'orange', 'red', 'green', 'purple'])
plt.ylim(0, 1)
plt.title('Visualización del Teorema de Bayes')
plt.ylabel('Probabilidades')

# Mostrar valores sobre las barras
for i, v in enumerate(values):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontweight='bold')

plt.show()
```
