###DISTRIBUCION NORMAL


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, kstest

# Datos de matrimonios por año de 2013 a 2023
matrimonios_por_anio = np.array([122, 136, 140, 113, 116, 94, 100, 95, 113, 117, 110])

# Calcular la media y la desviación estándar de los matrimonios
media_matrimonios = np.mean(matrimonios_por_anio)
desviacion_matrimonios = np.std(matrimonios_por_anio)

# Graficar el histograma de los matrimonios por año
plt.hist(matrimonios_por_anio, bins=5, density=True, alpha=0.6, color='g', label='Histograma')

# Crear la distribución normal con los parámetros calculados (media y desviación estándar)
x = np.linspace(min(matrimonios_por_anio) - 10, max(matrimonios_por_anio) + 10, 1000)
pdf = norm.pdf(x, media_matrimonios, desviacion_matrimonios)

# Superponer la curva normal en el histograma
plt.plot(x, pdf, color='red', label='Curva Normal Ajustada')

# Añadir etiquetas y título
plt.title('Histograma de Matrimonios con Curva Normal Ajustada')
plt.xlabel('Número de Matrimonios')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Realizar la prueba de Kolmogorov-Smirnov para evaluar el ajuste a la distribución normal
ks_statistic, ks_p_value = kstest(matrimonios_por_anio, 'norm', args=(media_matrimonios, desviacion_matrimonios))

# Mostrar los resultados de la prueba
print(f'Estadístico de Kolmogorov-Smirnov: {ks_statistic:.4f}')
print(f'Valor p de Kolmogorov-Smirnov: {ks_p_value:.4f}')

# Interpretación de la prueba
if ks_p_value > 0.05:
    print("Los datos se ajustan bien a la distribución normal (p > 0.05).")
else:
    print("Los datos no se ajustan bien a la distribución normal (p <= 0.05).")
```

###DISTRUBUCION EXPONENCIAL

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, kstest

# Datos de matrimonios por año de 2013 a 2023
matrimonios_por_anio = np.array([122, 136, 140, 113, 116, 94, 100, 95, 113, 117, 110])

# Calcular los intervalos entre los matrimonios (en años)
intervalos_entre_matrimonios = np.diff(np.arange(2013, 2024))

# Estimar el parámetro lambda (tasa de ocurrencia) de la distribución exponencial
media_intervalo = np.mean(intervalos_entre_matrimonios)
lambda_estimada = 1 / media_intervalo  # Tasa de ocurrencia

# Graficar el histograma de los intervalos
plt.hist(intervalos_entre_matrimonios, bins=6, density=True, alpha=0.6, color='g', label='Histograma de Intervalos')

# Crear la distribución exponencial con el parámetro lambda estimado
x = np.linspace(min(intervalos_entre_matrimonios) - 1, max(intervalos_entre_matrimonios) + 1, 1000)
pdf_exponencial = expon.pdf(x, scale=1/lambda_estimada)

# Superponer la curva exponencial en el histograma
plt.plot(x, pdf_exponencial, color='red', label=f'Curva Exponencial Ajustada (λ={lambda_estimada:.2f})')

# Añadir etiquetas y título
plt.title('Histograma de Intervalos entre Matrimonios con Curva Exponencial Ajustada')
plt.xlabel('Intervalos entre Matrimonios (años)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Realizar la prueba de Kolmogorov-Smirnov para evaluar el ajuste a la distribución exponencial
ks_statistic, ks_p_value = kstest(intervalos_entre_matrimonios, 'expon', args=(0, 1/lambda_estimada))

# Mostrar los resultados de la prueba
print(f'Estadístico de Kolmogorov-Smirnov: {ks_statistic:.4f}')
print(f'Valor p de Kolmogorov-Smirnov: {ks_p_value:.4f}')

# Interpretación de la prueba
if ks_p_value > 0.05:
    print("Los datos se ajustan bien a la distribución exponencial (p > 0.05).")
else:
    print("Los datos no se ajustan bien a la distribución exponencial (p <= 0.05).")

# Valor estimado de λ
print(f"Valor estimado de λ (tasa de ocurrencia): {lambda_estimada:.2f}")
```

###DISTRIBUCION CHI-CUADRADA

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos de matrimonios por año
anios = np.array([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
matrimonios = np.array([122, 136, 140, 113, 116, 94, 100, 95, 113, 117, 110])

# Crear variable categórica: Año Par o Impar
categoria_anio = np.where(anios % 2 == 0, "Par", "Impar")

data = pd.DataFrame({"Año": anios, "Matrimonios": matrimonios, "Categoría Año": categoria_anio})

# Construcción de tabla de contingencia
tabla_contingencia = pd.crosstab(data["Categoría Año"], data["Matrimonios"])

# Prueba de chi-cuadrado
chi2, p, dof, expected = stats.chi2_contingency(tabla_contingencia)

# Mostrar resultados
print("Tabla de contingencia:")
print(tabla_contingencia)
print(f"\nEstadístico chi-cuadrado: {chi2}")
print(f"p-valor: {p}")

# Gráfica de frecuencias observadas y esperadas
fig, ax = plt.subplots(figsize=(10, 5))
x_labels = tabla_contingencia.columns
width = 0.4

plt.bar(x_labels - width/2, tabla_contingencia.loc['Par'], width=width, label='Observados (Par)', alpha=0.7)
plt.bar(x_labels + width/2, tabla_contingencia.loc['Impar'], width=width, label='Observados (Impar)', alpha=0.7)
plt.xlabel("Cantidad de Matrimonios")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de Matrimonios por Año Par o Impar")
plt.legend()
plt.show()

# Interpretación
if p < 0.05:
    print("Existe una relación significativa entre la cantidad de matrimonios y la paridad del año.")
else:
    print("No hay evidencia suficiente para afirmar que exista relación entre las variables.")
```
